Introdução

Usando o csv do Titanic com a intenção de ver as diferenças de gêneros que embarcaram naquele fatídico dia. 

Usando o pacote de linguagem Pandas para manipulação de DataFrames , Numpy para realizar opreçãoes que envolva cálculos e Matplotlib ferramenta de visualização de gráficos.
     

```python
import pandas as pd 

import numpy as np

import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("titanic.csv")
df
```

```python
# realizando o processamento dos dados

def processamento_df(df):
    df  
    df.info() # Trazer informações do datafreme
    
    df.isnull() # Coferir valores nulos
    
    return df, df.info(), df.isnull # Retornar as informações solicitadas
             
```

```python
df, df.info(), df.isnull
```

```python
         
# Quantidade de mulheres a bordo
# Criando uma variável para chamr informações sobre a coluna Sex.

contagem_passageiras = df[df["Sex"].str.contains("female")]
contagem_passageiras
```

```python
# Quantidade de homens a bordo

contagem_passageiros = df[df["Sex"].str.contains("male")]
contagem_passageiros
```

```python
# Gerando um gráfico de linhas para visualização do quantitativo de mulheres a bordo por classe

import matplotlib.pyplot as plt
contagem_passageiras = contagem_passageiras['Pclass'].value_counts()
contagem_passageiras.plot(kind='bar')

plt.xlabel('Classe do Passageira')
plt.ylabel('Quantidade')
plt.title('Número de Passageiras por Classe')
```

```python

# Gerando um gráfico de linhas para visualização do quantitativo de homens a bordo por classe

contagem_passageiros = contagem_passageiros['Pclass'].value_counts()
contagem_passageiros.plot(kind='bar')

plt.xlabel('Classe do Passageiro')
plt.ylabel('Quantidade')
plt.title('Número de Passageiros por Classe')
```

```python
# Média de idade

genero = df.groupby("Sex")
media_genero = genero["PassengerId"].mean()
media_genero
```

```python
# Gerando um gráfico de linhas para visualização da média de passageiros por gênero

media_genero.plot.bar()
plt.title('Conatgem da média por gênero')

```

```python
# contagem total de passageiros por gênero, usando um gráfico de pizza

contagem_genero = genero["PassengerId"].count()
contagem_genero
```

```python
contagem_genero.plot.pie()
plt.title("Total de passageiros por gênero")
```
